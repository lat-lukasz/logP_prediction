### Scrapping from ZINC database logP value and SMILES string
Sterling and Irwin, J. Chem. Inf. Model, 2015 http://pubs.acs.org/doi/abs/10.1021/acs.jcim.5b00559. 

```python
import pandas as pd
import numpy as np
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
import re
import tqdm
import time
```

```python
def get_url(url):
    response = requests.get(url)
    if response.status_code == 200:
        print('Succesful Request')
        print(url)
    else:
        print('Request failed - status code: ', response.status_code)
        print(url)
    return response.text

def get_smiles_logP(subst_url):
    base_url = 'http://zinc.docking.org'
    bs = BeautifulSoup(get_url(base_url+subst_url), 'lxml')
    smiles = str(bs.find(id='substance-smiles-field')['value'])
    raw_logP = str(bs.find_all('td')[4])
    logP = re.sub('[^0-9\,\.]','', raw_logP).replace(',', '.')
    return smiles, logP
    
base_url = 'http://zinc.docking.org'
first_page_url = 'http://zinc.docking.org/substances/?page=1'

def get_ZINC_data(start,stop):
    df = pd.DataFrame(columns=['SMILES','logP'])
    for page in range(start,stop):
        page_url = 'http://zinc.docking.org/substances/?page={}'.format(page)
        bs = BeautifulSoup(get_url(page_url), 'lxml')
        for rows in bs.find_all('a', href=True):
            if 'substances/ZINC' in rows['href']:
                smiles, logp = get_smiles_logP(rows['href'])
                df = df.append({'SMILES': smiles,
                                    'logP': logp},ignore_index=True)
    return df
```

```python
x=[((int(str(x)+'0'))-10,int(str(x)+'0')) for x in range(1,1000)]
i=1
for x,y in x:
    try:
        data = get_ZINC_data(x,y)
        path = r'C:\Users\lukasz.latka\Desktop\projektu\logP_prediction_rdkit_scrapping\project\input\data{}.csv'.format(i)
        data.to_csv(path,index=False)
        i += 1
    except:
        pass
    
```
